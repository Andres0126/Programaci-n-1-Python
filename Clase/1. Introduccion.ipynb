# COMENTARIOS

A medida que los programas se hacen más grandes y complicados, se vuelven más difíciles de leer. Los lenguajes formales son densos, y a menudo es difícil mirar un trozo de código y averiguar qué está haciendo, o por qué. Por esta razón, es una buena idea añadir notas a tus programas para explicar en lenguaje natural lo que el programa está haciendo. Estas notas se llaman comentarios.

Un comentario en un programa de ordenador es un texto que está destinado sólo al lector humano - es completamente ignorado por el intérprete. En Python, el token # inicia un comentario. El resto de la línea se ignora. Aquí hay una nueva versión de ¡Hola, mundo!

````python
#---------------------------------------------------
# Es un ejemplo de la facilidad de utilziar comentarios!
# Cualquiera puede modificar o copiar este programa.
#---------------------------------------------------
#hola
print("Hola, MUNDO!")    # No es tan facil!
````

Observa que ningun comentario aparece, solo se imprime Hola, MUNDO!....

También notarás que hemos dejado una línea en blanco en el programa. Las líneas en blanco también son ignoradas por el intérprete.

Los comentarios y las líneas en blanco pueden hacer los programas más fáciles de entender. Utilízalas libremente.

En Python 3, que es el que utilizaremos, el operador

de división / produce un resultado en coma flotante (incluso si el resultado es un entero; 4/2 es 2.0). Si quieres una división truncada, que ignore el resto, puedes usar el operador // (por ejemplo, 5//2 es 2).

El operador módulo resulta ser sorprendentemente útil. Por ejemplo, puede comprobar si un número es divisible (*de forma exacta*) por otro: si x % y es cero, entonces x es divisible por y.

El operador % también puede extraer el dígito o los dígitos más a la derecha de un número entero. Por ejemplo, x % 10 da el dígito más a la derecha de x (en base 10). Del mismo modo, x % 100 da los dos últimos dígitos.

````python
# 125 divido 2
print(125 / 2)
print(125 // 2)  # extrae el cociente
# el operador módulo (%) devuelve el residuo de la división
print(125%2) # extrae el residuo

print(785/10)
print(785 % 10) # extrae el ultimo digito, es decir, el residuo de la division
print(78%10)
````

# VARIABLES

En ocasiones deseamos que el ordenador recuerde ciertos valores para usarlos más adelante, usando variables.

````python
#Se crea un objeto entero con el valor 3.14 en memoria. La variable pi apunta a ese objeto.
pi = 3.1416
print(pi)
radio =2
area = pi*radio**2
print("Area del circulo:", area)

print(id(pi))
print(id(area))
print(id(pi) == id(area))

radio =1.5
area = pi*radio**2
print(f'Area del circulo= {area}')


a = 1
b = 1
print(id(a) )
print(id(b) )

texto= "hola"
print(type(texto))
print(id(texto))
````

El acto de dar valor a una variable se denomina asignación. Al asignar un valor a una variable que no existía, Python reserva un espacio en la memoria, almacena el valor en él y crea una asociación entre el nombre de la variable y la dirección de memoria de dicho espacio.
