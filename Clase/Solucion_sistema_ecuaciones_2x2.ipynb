Solución de un sistema de ecuaciones de dos incognitas dos ecuaciones usanod el método de igualación.

a1x + b1y = c1
a2x + b2y = c2

```python
# A continuacion programamos la linea 1 y 2 del seudocodigo
a1 = float(input("Ingrese el valor de la constante a1: ")) # en esta linea se define y se lee la variable a1
b1 = float(input("Ingrese el valor de la constante b1: ")) # en esta linea se define y se lee la variable b1
c1 = float(input("Ingrese el valor de la constante c1: ")) # en esta linea se define y se lee la variable c1

a2 = float(input("Ingrese el valor de la constante a2: ")) # en esta linea se define y se lee la variable a1
b2 = float(input("Ingrese el valor de la constante b2: ")) # en esta linea se define y se lee la variable b1
c2 = float(input("Ingrese el valor de la constante c2: ")) # en esta linea se define y se lee la variable c1
x = 0; y=0; # en esta parte se inicializa y se define las variables x y y. Python las define como el tipo de variable entero,
#porque 0 es un entero. Python es un lenguaje de tipado dinámico, lo que significa
# que determina el tipo de una variable en tiempo de ejecución, basado en el valor que le asignes.


if (a1 ==0) & (b1 == 0):
  print("No es un sistema de ecuaciones!!... \n Verifica el valor de las constantes a1 y b1!!..")
elif (a2 ==0) & (b2 == 0):
  print("No es un sistema de ecuaciones!!... \n Verifica el valor de las constantes a2 y b2!!..")
else:
  # dentro de este if, programamos la linea 3 del seudocodigo en adelante
  print("Ha ingresado correctamente un sistema de ecucaciones")
  if (a1!=0) & (a2!=0):
     if (b1!=0) | (b2!=0): # aca es un "o" y no un "y"
        y = (c2/a2 - c1/a1)/(b2/a2 - b1/a1)
        x =(c1-b1*y)/a1
     else:
        print("El sistema no tiene solución dado que son dos rectas verticales paralelas")
        print(f'recta 1: x={c1/a1}, recta 2: x={c2/a2}')

  elif (a1==0) & (a2==0):
    print("El sistema no tiene solución dado que son dos rectas horizontales paralelas")
    print(f'recta 1: y={c1/b1}, recta 2: y={c2/b2}')

  elif (a1==0):
        y = c1/b1
        x = (c2 - b2*y)/a2
  elif (a2==0):
        y = c2/b2
        x = (c1 - b1*y)/a1
  print(f"La solución del sistema de ecuaciones: x ={x} y y={y}")
```
